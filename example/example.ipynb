# Use POV-Ray inside a jupyter notebook!

```povray
#include "colors.inc"
camera {
    location <1, 5, -6>
    look_at  <1, 0, 0>
}
light_source {
    <10, 20, -10>
    color rgb 1
}
light_source {
    <-5, 0, -5>
    color rgb 1
}
background {
    color Black
}
plane { <0, 1, 0>, -0.2
    pigment {
        checker color White, color Black
    }
}
text {
    ttf "/usr/share/fonts/liberation/LiberationSans-Regular.ttf"
    "jansjupyterkernel", 0.25, 0
    pigment {
        color Red
    }
    translate -3*x
}
```

```povray
#include "colors.inc"
background { color Black }
camera {
    location <0, 2, -3>
    look_at  <0, 1,  2>
}
sphere {
    <0, 1, 2>, 2
    texture {
          pigment { color Red }
    }
}
light_source { <2, 4, -3> color White}
```
